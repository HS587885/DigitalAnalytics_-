 1. 허수아비의 체력은 100이다 
 2. 플레이어의 공격을 수치로 입력 받습니다.
 3. 허수아비의 체력을 넘는 공격을 준다면 허수아비가 쓰러집니다.
 4. 허수아비의 체력을 밑도는 공격을 준다면 허수아비는 쓰러지지 않습니다.
 5. 공격력이 10정도로 모자라다면 지나가던 개가 "아쉽네" 라고 말합니다.


```python
scarecrowhp = 100
damage = int(input("플레이어의 공격력"))

if damage > scarecrowhp:
  msg = "허수아비가 쓰러졌다!"
elif damage < scarecrowhp:
  msg = "공격력이 약해 허수아비가 쓰러지지 않습니다" 
  if scarecrowhp - damage  <= 10:
      msg = "지나가던 개가 말함: '아쉽네~~^^ '"
else:
  msg = "허수아비는 쓰러지지 않았습니다"

print(msg)
```

# 응용문제 1 : 임금계산 프로그램 학습하기

문제이해하기

사용자에게 주당 근무시간과 시간당 임금을 물어본다. 주당 근무시간이
40시간을 넘으면 초과근무시간은 1.5배 임금을 지급해야 한다. 
이번 주에 받는 총 임금을 계산하는 프로그램을 작성하시오.

```python
worktime = int(input("근무시간을 입력하세요."))
hourperwage = 10000
time_standard = 40 


stardard_pay = (time_standard * hourperwage) # 40시간 * 10000원
general_pay = (worktime * hourperwage) # input()값 * 10000원
overwork_pay = ((worktime - time_standard) * hourperwage) * 1.5

if worktime > 40:
  total_pay = stardard_pay + overwork_pay
else:
  total_pay = general_pay

print(f"""
근무시간: {worktime}
시간당 임금: {hourperwage}
total_pay: {int(total_pay)}

""")

```

# 응용문제 3 : 자동판매기 프로그램 학습하기

문제이해하기

자동판매기는 사용자로부터 투입한 돈을 입력 받는다. 물건 값은 100원 단위라고 가정한다. 
프로그램은 잔돈을 계산하여 출력한다.
- 음료와 판매금액은 딕셔너리 구조를 사용한다.
- 자판기는 동전 500원, 100원, 50원짜리만 가지고 있다.

```python
print("-" * 50)
print("""
1: coke       500
2: coffee     600
3: cider      500
4: water      400
""")
print("-"* 50)


vending_machine = {'coke':500, 'coffee': 500, 'cider':500, 'water':400}
vm = vending_machine
inserted_coin = int(input("투입한 돈: "))
number_chosen = int(input("번호 선택: "))


if number_chosen == 1:
  if inserted_coin < vm['coke']:
    print("잔액이 부족합니다")
  else:
    change = inserted_coin - vm['coke']
    five_hun = change // 500
    hun =  (change - (five_hun * 500)) // 100
    fifty = (change  - ((hun * 100) + (five_hun * 500))) // 50
    left = (change  - ((hun * 100) + (five_hun * 500))) % 50


elif number_chosen == 2:
   if inserted_coin < vm['coffee']:
     print("잔액이 부족합니다")
   else:
    change = inserted_coin - vm['coffee']
    five_hun = change // 500
    hun =  (change - (five_hun * 500)) // 100
    fifty = (change  - ((hun * 100) + (five_hun * 500))) // 50
    left = (change  - ((hun * 100) + (five_hun * 500))) % 50
    

elif number_chosen == 3:
  if inserted_coin < vm['cider']:
    print("잔액이 부족합니다")
  else:
    change = inserted_coin - vm['cider']
    five_hun = change // 500
    hun =  (change - (five_hun * 500)) // 100
    fifty = (change  - ((hun * 100) + (five_hun * 500))) // 50
    left = (change  - ((hun * 100) + (five_hun * 500))) % 50
 
elif number_chosen == 4:
  if inserted_coin < vm['water']:
    print("잔액이 부족합니다")
  else:
    change = inserted_coin - vm['water']
    five_hun = change // 500
    hun =  (change - (five_hun * 500)) // 100
    fifty = (change  - ((hun * 100) + (five_hun * 500))) // 50
    left = (change  - ((hun * 100) + (five_hun * 500))) % 50


print(f"""
거스름돈: {change}원
=============================
500원 동전: {five_hun}개
100원 동전: {hun}개
50원 동전: {fifty}개
바꾸지 못한 돈: {left}원    
    
    """)
  
  
```

# 응용문제 4 : 지불금액 계산 프로그램 학습하기

문제이해하기

다음은 구매 금액에 따른 할인율을 적용하여 지불 금액을 계산하는 프로그램이다. 
단, 금액에 따른 할인율은 아래 표와 같다.


image.png

```python
purchased_price = int(input("구매 금액은? "))
print('-'* 50)
ptp = purchased_price

if ptp < 10000:
  discount_rate = 0
  dr = "0%"
  discount_price = 0
  price_to_pay = purchased_price 
elif ptp >= 10000 and  ptp < 50000:
  discount_rate = 0.05
  dr = "5%"
  discount_price = ptp * discount_rate
  price_to_pay = ptp - discount_price

elif ptp >= 50000 and  ptp < 200000:
  discount_rate = 0.1
  dr = "10%"
  discount_price = ptp * discount_rate
  price_to_pay = ptp - discount_price  

else: 
  discount_rate = 0.2
  dr = "20%"
  discount_price = ptp * discount_rate
  price_to_pay = ptp - discount_price 


print(f"""
구매금액: {ptp}
할인율: {dr}
할인금액: {int(discount_price)}
지불금액: {int(price_to_pay)}
""")
print('-'* 50)
```
