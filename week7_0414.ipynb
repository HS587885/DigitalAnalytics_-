```python
import random

#함수정의

def getNum():
  r = random.randint(1,45)
  return r

#메인코드
lotto=[]

print("**로또 프로그램을 시작합니다.**")

while True:
  n = getNum()
  if lotto.count(n) == 0:
    lotto.append(n)
  if len(lotto) == 6: break

print('추첨된 로또 번호:', end = ' ')
lotto.sort()

for i in range(6):
  print(f'{lotto[i]}', end = ' ')
```

```python
filename = '/content/drive/MyDrive/data/score.txt'
f = open(filename, encoding = 'utf-8')
data = f.readlines()
df = []
for line in data:
  df.append(line.rstrip('\n').split(','))


def average_grade(df):
  average = []
  grade= []
  for i in range(len(df)):
    average_cal = (int(df[i][1]) + int(df[i][2]) + int(df[i][3])) / 3
    average.append(average_cal)
  for j in average:
    if j  >= 90:
      grade_cal = 'A'
      grade.append(grade_cal)
    elif j >= 80:
      grade_cal = 'B'
      grade.append(grade_cal)
    else:
      grade_cal = 'C'
      grade.append(grade_cal)
  return average, grade

def main():
  cols = ["이름","국어","영어","수학","평균", "학점"]
  average_grade(df)
  print(f"{cols[0]},{cols[1]},{cols[2]},{cols[3]},{cols[4]},{cols[5]}")
  print("-" * 25)
  for item in range(len(cols)):
    print(f"{df[item][0]},{df[item][1]},{df[item][2]},{df[item][3]},{round(average[item], 2)},{grade[item]}")
 
main()
```

```python
import pandas as pd
filename = '/content/drive/MyDrive/data/score.txt'
cols = ["이름","국어","영어","수학"]
f = open(filename, encoding = 'utf-8')
data = f.readlines()
df = []
for line in data:
  df.append(line.rstrip('\n').split(','))

data = pd.DataFrame(df, columns = cols)
data["국어"] = data["국어"].astype('int')
data["영어"] = data["영어"].astype('int')
data["수학"] = data["수학"].astype('int')
data['평균'] = (data['국어'] + data["영어"] + data['수학']) / 3
#data["평균"] = data["평균"].astype('int')
grade = list(data["평균"])

grade_list = []
for i in grade:
  if i >= 90:
    grade_list.append("A")
  elif i >= 80:
    grade_list.append("B")
  else:
    grade_list.append("C")
data["학점"] = grade_list
data
```

```python
import random as rd
filename = '/content/drive/MyDrive/word.txt'
print("타자게임 시작")
f = open(filename, encoding = 'utf-8')
data = f.readlines()
  


def tajaGame(data):
  switch = True
  while switch:
    for line in rd.sample(data, 1):
      line = line.rstrip('\n')
      print(line)
      typing = input()
      if line == typing:
        print("맞음!!")
        continue
      else:
        if len(typing) < 1:
          switch = False
        else:
          print("오타! 다시도전")
        
tajaGame(data)
```
