```python
filename = '/content/drive/MyDrive/data/score.txt'
f = open(filename, encoding = 'utf-8')
data = f.readlines()
df = []
for line in data:
  df.append(line.rstrip('\n').split(','))


def average_grade(df):
  average = []
  grade= []
  for i in range(len(df)):
    average_cal = (int(df[i][1]) + int(df[i][2]) + int(df[i][3])) / 3
    average.append(average_cal)
  for j in average:
    if j  >= 90:
      grade_cal = 'A'
      grade.append(grade_cal)
    elif j >= 80:
      grade_cal = 'B'
      grade.append(grade_cal)
    else:
      grade_cal = 'C'
      grade.append(grade_cal)
  return average, grade

def main():
  cols = ["","","","","", ""]
  average_grade(df)
  print(f"{cols[0]},{cols[1]},{cols[2]},{cols[3]},{cols[4]},{cols[5]}")
  print("-" * 25)
  for item in range(len(cols)):
    print(f"{df[item][0]},{df[item][1]},{df[item][2]},{df[item][3]},{round(average[item], 2)},{grade[item]}")
 
main()
```

```python
import pandas as pd
filename = '/content/drive/MyDrive/data/score.txt'
cols = ["","","",""]
f = open(filename, encoding = 'utf-8')
data = f.readlines()
df = []
for line in data:
  df.append(line.rstrip('\n').split(','))

data = pd.DataFrame(df, columns = cols)
data[""] = data[""].astype('int')
data[""] = data[""].astype('int')
data[""] = data[""].astype('int')
data[''] = (data[''] + data[""] + data['']) / 3
#data[""] = data[""].astype('int')
grade = list(data[""])

grade_list = []
for i in grade:
  if i >= 90:
    grade_list.append("A")
  elif i >= 80:
    grade_list.append("B")
  else:
    grade_list.append("C")
data[""] = grade_list
data
```

```python
import random as rd
filename = '/content/drive/MyDrive/word.txt'
print(" ")
f = open(filename, encoding = 'utf-8')
data = f.readlines()
  


def tajaGame(data):
  switch = True
  while switch:
    for line in rd.sample(data, 1):
      line = line.rstrip('\n')
      print(line)
      typing = input()
      if line == typing:
        print("!!")
        continue
      else:
        if len(typing) < 1:
          switch = False
        else:
          print("! ")
        
tajaGame(data)
```
