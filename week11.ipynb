```python
class Person:
  def __init__(self,name, number):
    self.name= name
    self.number = number
```

```python
class Student(Person):
  UNDERGRADUATE = 0
  POSTGRADUATE = 1

  def __init__(self,name,number,studentType):
    super().__init__(name, number) #부모클래스 호출
    self.studentType = studentType
    self.classes = []
    self.gpa = 0

  def enrollCourse(self, c= ""):
     self.classes.append(c)

  def __str__(self):
    super().__str__()
    result = f' 이름:  {self.name} \n 주민번호: {self.number} \n 수강과목: {self.classes} \n 평점: {self.gpa}'
    return result
    

 
class Teacher(Person):
 
  def __init__(self,name,number):
    super().__init__(name, number) #부모클래스 호출
    self.courses = []
    self.salary = 3000000

  def assignTeaching(self, c= ""):  
   
    self.courses.append(c)
    
  
  def __str__(self):
    super().__str__()
    result = f' 이름:  {self.name} \n 주민번호: {self.number} \n 강의과목: {self.courses} \n 월급: {self.salary}'
    return result
    

   
```

```python
class Ticket:
  def __init__(self, number, price):
    self.number = number
    self.price = price

  def getNumber(self):
    return self.number

  def getPrice(self):
    return self.price
  
  def setPrice(self, new_price):
    self.price = new_price
    return self.price

  def __str__(self):
    result = f'입장권 번호:{self.number},가격:{self.price}'
    return result

```

```python
class WalkupTicket(Ticket):
  def __init__(self,number, price, payBYcredit):
    super().__init__(number, price) #부모클래스 호출
    self.payBYcredit = bool(payBYcredit)
    
  def getPayByCredit(self):
    return self.payBYcredit

  def getPrice(self):
    if self.getPayByCredit() == True:
      self.price = self.price + (self.price * 0.05)
    else:
      self.price
    return self.price

  def __str__(self):
     super().__str__()
     result = f' 번호: {self.number},  신용카드 결제: {self.getPayByCredit()}, 가격: {int(self.getPrice())}'
     return result


class AdvanceTicket(Ticket):
  def __init__(self,number,price, advancedays):
    super().__init__(number, price) #부모클래스 호출
    self.advancedays = advancedays
  
  def getAdvancedays(self):
    return self.advancedays  

  def getPrice(self):
    if self.getAdvancedays() >= 30:
      self.price = self.price * 0.5
    elif self.getAdvancedays() < 29 and self.getAdvancedays() >= 10:
      self.price = self.price * 0.8
    else:
      self.price = self.price * 0.9
    return self.price

    

  def __str__(self):
     super().__str__()
     result = f' 번호: {self.number},  사전 예약일수: {self.getAdvancedays()}, 가격: {int(self.getPrice())}'
     return result
```
