```python
class BankAccount:
  #생성자
  def __init__(self):
    self.balance = 0
  #출금메소드
  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -=amount
      print(f'통장에 {amount}원 출금됨')
    else:
      print("잔액부족")

  def deposit(self, amount):
    self.balance+=amount
    print(f'통장에 {amount}원 입금됨')
  
  def __str__(self):
    return f'현재잔액: {self.balance}'
```

```python
#부모클래스 = 슈퍼클래스 = 기저(base)클래스 = 상위클래스
class Person():
  #멤버변수
  name=''
  age=0
  #생성자
  def __init__(self,name,age):
    self.name=name
    self.age = age

  #일반메소드
  def setName(self,name):
    self.name=name

  def setAge(self,age):
    self.age=age
  
  def work(self):
    print('일은 한다.')
  #출력메소드
  def show(self):
    print(f'{self.name},{self.age}')
```

```python
#자식클래스 = 하위클래스=유도클래스
class Student(Person):
  def __init__(self,name,age,dept):
    super().__init__(name,age) #부모클래스 호출
    self.dept = dept

  def work(self):
    super().work()
    print('공부를 한다.')
  def show(self):
    super().show()
    print(f'{self.dept}')
```

# 실습예제 2 : 직원과 매니저

```python
class Employee:
  def __init__(self,name,salary):
    self.name=name
    self.salary=salary
  def getSalary(self):
    return self.salary
  def __str__(self):
    result=f'이름:{self.name}, 월급:{self.salary}'
    return result
  
class Manager(Employee):
  def __init__(self,name,salary,bonus):
    super().__init__(name,salary)
    self.bonus=bonus
  def getSalary(self):
    salary=super().getSalary()
    salary=salary+self.bonus
    return salary
  def __str__(self):
    result=f'{super().__str__()}, 보너스:{self.bonus}, 전체급여:{self.getSalary()}'
    return result

e1=Employee('구글신',200000)
print(e1)
m1=Manager('자바신',2000000,100000)
print(m1)
```

# 응용문제1 : Person 클래스

```python
class Person:
    def __init__(self, name, number):
        self.name = name
        self.number = number
    def __str__(self):
      result=f'이름 :{self.name}\n주민번호:{self.number}'
      return result

class Student(Person):
    UNDERGRADUATE=0 # 재학생
    POSTGRADUATE=1 # 졸업생

    def __init__(self, name, number, studentType ):
        super().__init__(name, number)
        self.studentType = studentType
        self.gpa=0.0
        self.classes = []

    def enrollCourse(self, course):
        self.classes.append(course)

    def __str__(self):
        result=super().__str__()
        result+=f'\n수강과목:{self.classes}\n평점:{self.gpa}'
        return result

class Teacher(Person):
    def __init__(self, name, number):
        super().__init__(name, number)
        self.courses = []
        self.salary=3000000

    def assignTeaching(self, course):
        self.courses.append(course)

    def __str__(self):
        result=super().__str__()
        result+=f'\n강의과목:{self.courses}\n월급:{self.salary}'
        return result
```

```python
class Ticket:
  def  __init__(self,number,price):
   self.number = number
   self.price = price
   
  def getNumber(self):
    return self.number

  def getPrice(self):
    return self.price
  
  def setPrice(self,price):
    self.price = price
  
  def __str__(self):
    result = f'번호 : {self.getNumber()} \n가격 : {self.getPrice()}'
    return result
```

```python
class WalkupTicket(Ticket):
  def __init__(self,number,price,payBYcredit):
    super().__init__(number,price)
    self.payBYcredit = payBYcredit

  def getPrice(self):
    return super().getPrice()

  def getPayByCredit(self):
    if self.payBYcredit == 'TRUE':
      price = int(self.price * 1.05)
    else:
      price = self.price
    return int(price)
  
  def __str__(self):
    result = f'번호 : {self.number} 신용카드 결제 : {self.payBYcredit}  가격 : {self.getPayByCredit()}'
    return result
```

```python
a1 = WalkupTicket(101,100000,'TRUE')
a2 = WalkupTicket(101,100000,'FALSE')
print(a1)
print(a2)
```

```python
#예약 입장권


class AdvanceTicket(Ticket):
  def __init__(self,number,advancedays,price):
    super().__init__(number,price)
    self.advancedays = advancedays

  def getPrice(self):
    return super().getPrice()

  def getAdvancedays(self):
    days = self.advancedays
    if days <=4:
      rate = 1
    elif days <=9:
      rate = 0.9
    elif days <29:
      rate = 0.8
    else:
      rate = 0.5
    pay = rate*self.price
    return int(pay)

  def __str__(self):
    result = f'번호 {self.number} 사전 예약일수 : {self.advancedays} 가격 : {self.getAdvancedays()}'
    return result
```

```python
a1 = WalkupTicket(101,100000,'TRUE')
a2 = WalkupTicket(101,100000,'FALSE')
print(a1)
print(a2)
```

```python
b1 = AdvanceTicket(201,30,100000)
b2 = AdvanceTicket(301,20,100000)
b3 = AdvanceTicket(401,8,100000)
b4 = AdvanceTicket(501,3,100000)
print(b1)
print(b2)
print(b3)
print(b4)
```

- 넘파이 배열을 사용하면 배열 간에 수학적인 연산을 적용
- 고급 연산자와 풍부한 함수를 제공한다.

```python
mid=[10,20,30]
final=[10,20,30]
```

```python
# 리스트
n = len(mid)
total, avg=[], []

for i in range(n):
  total.append(mid[i] + final[i])
  avg.append(total[i]/2)
print(total)
print(avg)
```

```python
# 2) 필요없는 열을 삭제한다.
vaccine.drop(['day1sum','day1rate','day2sum','day2rate'], axis=1,inplace=True)
vaccine


#필요없는 컬럼 삭제 후 출력 
del_lst = ["day1sum",	"day1rate",	"day2num",	"day2sum",	"day2rate"]
for i in del_lst:
  vaccine.drop([i], axis = 1, inplace = True)
vaccine.head()
```

```python
# 3) 기초통계량
vaccine.describe()
```

```python
vaccine['date']=pd.to_datetime(vaccine['date'])# astype~
vaccine.dtypes
```

```python
vaccine['date'] = pd.to_datetime(vaccine['date'], format='%Y.%m.%d') 
```

```python
# 4) 새로운 열('월')생성
vaccine['month']=vaccine['date'].dt.month
vaccine
```

```python
# 접종일의 '월'로 새로운 열 'month' 생성후 월별 평균 출력

means=vaccine.groupby('month')[['day1num','day2num']].mean()
means
```

## 결측데이터 최빈값으로 대체하기

```python
# 탑승항구(Embarked)는 최빈값으로 대체
# 최빈값 찾기
titanic['Embarked'].value_counts()
```

```python
# 최빈값으로 대체하기
titanic['Embarked'].fillna('S',inplace=True)
```

```python
titanic.isnull().sum()
```

```python
# mode()함수 이용
f=titanic['Embarked'].mode()[0]
titanic['Embarked'].fillna(f,inplace=True)
```

```python
# 나이(Age)는 평균으로 대체하기
avg=titanic['Age'].mean() # max(),min(),sum(),count()
titanic['Age'].fillna(avg,inplace=True)
```

```python
# 결측데이터 평균으로 대체하기
titanic['Age'].fillna(0,inplace=True)
```
