## for 리스트의 활용2
 - 힉생 5명의 성적이 리스트에 저장 60점이상이면 합격 아니면 불합격

```python
grade = [80, 90, 74, 88,100]

for k, g in enumerate(grade):
  
  if g >= 60:
    result = '합격'
  else:
    result = '불합격'
  
  print(f'{k + 1}번 학생의 {g} 점수는 {result}입니다')

```

# 응용문제 1 : 커피자판기 프로그램


```python
mlist = ['뜨아', '아아', '라떼', '모카']
plist = [2000, 3000, 3500, 4000]

print('** 자판기 판매 메뉴**')
for i in range(0, len(mlist)):
    print(f'{i + 1} : {mlist[i]} {plist[i]}')

money = int(input("돈을 투입하세요 : "))

while True:
    
    choice = int(input("메뉴 선택 (종료: 0)"))
    temp = money
    if choice != 0:
        if choice == 1 and money  >= plist[0]:
            print(f'{mlist[0]} 구입완료')
            money = temp - plist[0]      
            if money <= 0:
                print("잔액부족")
                print(f' 잔액 {money} ')
                
        elif choice == 2 and money >= plist[1]:
            print(f'{mlist[1]} 구입완료')
            money = temp - plist[1]      
            if money <= 0:
                print("잔액부족")
                print(f' 잔액 {money} ')
                
        elif choice == 3 and money  >= plist[2]:
            print(f'{mlist[2]} 구입완료')
            money = temp - plist[2]      
            if money <= 0:
                print("잔액부족")
                print(f' 잔액 {money} ')
    
        elif choice == 4 and money >= plist[3] :
            print(f'{mlist[3]} 구입완료')
            money = temp - plist[3]      
            if money <= 0:
                print("잔액부족")
                print(f' 잔액 {money} ')           
        else: 
            print('잔액부족')
        change = money 
        print(f' 잔액 {change}')
    
        if change < 1999 or change < 0:
            print("잔액이 부족합니다")    
            print(f'자판기 종료, 잔액 {change} 반환 ')
            break
    else:
        print(f' 자판기 종료, 잔액 {money} 반환 ')
        break
    
```
