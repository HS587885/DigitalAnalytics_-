```python
# 조건에 맞는 데이터 추출
# 평균기온이 30도 이상인 데이터 추출
w = df['평균기온']>=30
df[w]
```

```python
# 평균기온이 20도 이상이고 30도 이하인 데이터 추출(&, |, ~)
w = (df['평균기온']>= 20) & (df['평균기온']<=30)
df[w]
```

```python
# 가장 더웠던 날의 모든 정보 추출
w= df['최고기온']==df['최고기온'].max()
df[w]
```

```python
# 일교차가 16도 이상인 데이터
w = df['최고기온'] - df['최저기온'] >= 16
df[w]
```
