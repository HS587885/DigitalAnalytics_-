```python
class BankAccount:
  #생성자
  def __init__(self):
    self.balance = 0
  #출금메소드
  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -=amount
      print(f'통장에 {amount}원 출금됨')
    else:
      print("잔액부족")

  def deposit(self, amount):
    self.balance+=amount
    print(f'통장에 {amount}원 입금됨')
  
  def __str__(self):
    return f'현재잔액: {self.balance}'
```

```python
#테스트
b1 = BankAccount()
b1.deposit(10000)
print(b1)
b1.withdraw(5000)
print(b1)
b1.withdraw(6000)
print(b1)
```

```python
class Car:
  def __init__(self,model, color, price):
    self.model = model
    self.color = color
    self.price = price
    
  def __str__(self):
    result = f'model:{self.model}, color:{self.color}, price:{self.price}'
    return result

  class ElectricCar():
    def __init__(self, model, color, price, batterySize):
      self.model=model
      self.color=color
      self.price=price
      self.batterySize = batterySize
    
    def __str__(self):
      result = f'model: {self.model}, color:{self.color}, price:{self.price}, batterySize: {self.batterySize}'
      return result
```

```python
class ElectricCar(Car):
  def __init__(self,model,color,price,batterySize):
    super().__init__(model,color, price) #부모생성출호출
    self.batterySize = batterySize

  def __str__(self):
    result=super().__str__() #부모 메소드 호출
    result += f'batterySize: {self.batterySize}'
    return result

    
```

-학생 클래스 : 이름, 나이, 학과, 일work

#상속의 장점
- 클래스 사이의 멤버 중복 선언 불필요
- 필드(멤버 변수)와 메소드 재사용으로 클래스 간결화
- 클래스 간 계층적 분류 및 관리

```python
#부모클래스 = 슈퍼클래스 = 기저(base)클래스 = 상위클래스
class Person():
  #멤버변수
  name=''
  age=0
  #생성자
  def __init__(self,name,age):
    self.name=name
    self.age = age

  #일반메소드
  def setName(self,name):
    self.name=name

  def setAge(self,age):
    self.age=age
  
  def work(self):
    print('일은 한다.')
  #출력메소드
  def show(self):
    print(f'{self.name},{self.age}')
```

```python
#자식클래스 = 하위클래스=유도클래스
class Student(Person):
  def __init__(self,name,age,dept):
    super().__init__(name,age) #부모클래스 호출
    self.dept = dept

  def work(self):
    super().work()
    print('공부를 한다.')
  def show(self):
    super().show()
    print(f'{self.dept}')
```

```python
#테스트
p1 = Person('성형훈',20)
p1.show()
```

```python
s1=Student('성형훈', 20, 'DA')
s1.show()
s1.setName('Tony')
s1.show()
s1.work()
```

```python
class Employee:
  def __init__(self, name, salary):
    self.name =name
    self.salary = salary

  def getSalary(self):
    return self.salary

  def __str__(self):
    result = f'이름:{self.name}, 월급:{self.salary}'
    return result


class Manager(Employee):
  def __init__(self, name, salary, bonus):
    super().__init__(name, salary)
    self.bonus=bonus

def getSalary(self):
  salary = self.salary + self.bonus
  return salary

def __str__(self):
  result = f'이름:{self.name},월급:{self.salary},보너스:{self.bonus}'
  return result
```

TV 클래스를 정의하고 객체를 생성한 후 테스트를 해보자
- 채널
- 볼륨
- 전원상태
- turnOn()
- turnOff()
- setChannel(channel) -채널을 변경한다
- setVolume(Volume) - 볼륨을 변경한다 

```python
class Tv:
  def __init__(self,channel,volume,condition):
    self.channel = channel
    self.volume = volume
    self.condition = condition
  
  def turnOn(self):
    condition = self.condition
    if condition == 1:
      msg = '이미켜짐'
    else:
      condition = 1
      msg = '켜짐'
    print(msg)
  
  def turnOff(self):
    condition = self.condition
    if condition == 0:
      msg = '이미 꺼짐'
    else:
      condition =0
      msg = '꺼짐'
    print(msg)

  def setChanel(self,channel):
      self.channel=channel
    
  def setVolume(self,volume):
      self.volume=volume
  def getChanel(self):
    return self.channel
  def getVolume(self):
    return self.volume
  def __str__(self):
      result=f'TV의 채널는 {self.getChanel()} TV의 볼륨은 {self.getVolume()}입니다.'
      return result
```

```python
tv1 = Tv(10,11,1)
# 채널변경
tv1.setChanel(15)
print(tv1)
# 볼륨변경
```
