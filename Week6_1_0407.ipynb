```python
coffee={'뜨아':2000,'아아':3000,'라떼':3500,'모카':4000}


# 메뉴출력기능
def menu():
  print('** 자판기 판매 메뉴**')
  order = 0
  for key in coffee:
      order += 1
      print(f'{order} : {key} {coffee[key]}')

# 동전교환기능 
def counter(money):
    five_hun = money // 500
    hun =  (money - (five_hun * 500)) // 100
    fifty = (money  - ((hun * 100) + (five_hun * 500))) // 50
    left = (money  - ((hun * 100) + (five_hun * 500))) % 50
    print(f""" 거스름돈: {money}원
        =============================
        500원 동전: {five_hun}개
        100원 동전: {hun}개
        50원 동전: {fifty}개
        바꾸지 못한 돈: {left}원    

            """)
        
# 계산 기능    

def cal_cal():
  menu()
  money = int(input("돈을 투입하세요 : "))
  while True:  
      choice = int(input("메뉴 선택 (종료: 0)"))
      temp = money
      if choice != 0:
          if choice == 1 and money >= coffee['뜨아']:
              print('뜨아 구입완료')
              money = temp - coffee['뜨아']      
              if money <= 0:
                  print("잔액부족")
                  print(f' 잔액 {money} ')
                  
          elif choice == 2 and money >= coffee['아아']:
              print('아아 구입완료')
              money = temp - coffee['아아']      
              if money <= 0:
                  print("잔액부족")
                  print(f' 잔액 {money} ')
                  
          elif choice == 3 and money  >= coffee['라떼']:
              print('라뗴 구입완료')
              money = temp - coffee['라떼']      
              if money <= 0:
                  print("잔액부족")
                  print(f' 잔액 {money} ')
      
          elif choice == 4 and money >= coffee['모카']:
              print('모카 구입완료')
              money = temp - coffee['모카']      
              if money <= 0:
                  print("잔액부족")
                  print(f' 잔액 {money} ')           
          else: 
              print('잔액부족')
          change = money 
          print(f' 잔액 {change}')
      
          if change < 1999 or change < 0:
              print("잔액이 부족합니다")    
              print(f'자판기 종료, 잔액 {change} 반환 ')
              counter(money)
              break
      else:
          print(f' 자판기 종료, 잔액 {money} 반환 ')
          counter(money)
          break

    
```

#응용문제 5

인공지능은 입력 데이터로 학습을 시켜야 한다. 입력데이터는 training_data를 사용, 2차원 리스트로 구성되어 있다. flower_counts()에서 입력데이터에 대한 꽃들의 개수를 카운트하여 그 정보를 딕셔너리 형식으로 저장한 후 출력해보자

컬렉션자료형(리스트, 딕셔너리)

함수정의와 호출

제어문

```python
training_data=[['노랑', 2, '개나리'],
           ['흰색', 5, '벚꽃'],
           ['노랑', 5, '개나리'],
           ['흰색', 2, '목련'],
           ['분홍색', 2, '진달래']
             ]
fl_dict = {}

def flower_counts(training_data):
  name_list = []
  td = training_data
  td1 = sum(td , [])
  for i in range(len(td)): 
    name_list.append(td[i][-1])
  for m in name_list:
    fl_dict[m] = td1.count(m)
  return fl_dict
  
  
```

```python
flower_counts(training_data)
```
