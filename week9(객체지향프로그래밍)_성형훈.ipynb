## 실습예제: 계산기 프로그램

```python
class Calculator():
  def __init__(self, n1, n2):
    self.num1 = n1
    self.num2 = n2

  def add(self):
    result = self.num1 + self.num2
    return result

  def sub(self):
    result = self.num1 - self.num2
    return result
```

```python
# 객체생성

c1 = Calculator(10, 20)
print(c1)
```

## 도전문제 1: 자동차 클래스 정의 및 객체 활용

```python
class Car:
  #생성자
  def __init__(self, color, speed):
    self.color = color
    self.speed = speed

  #메소드
  def upSpeed(self, value):
    self.speed+=value

  def downSpeed(self, value):
    self.speed-=value

  def printInfo(self):
    print(f'color: {self.color}, speed : {self.speed}')

  def __str__(self): #print()함수를 통해 객체의 정보를 출력 자동호출
    msg = f'color: {self.color}, speed : {self.speed}'
    return msg
```

```python
class Student:
  def __init__(self, name, kor, eng, math):
    self.name = name
    self.kor = kor
    self.eng = eng
    self.math =math

  def getSum(self):
    total = self.kor + self.eng + self.math
    return total

  def getAvg(self):
    average = self.getSum()/3
    return average

  def getGrade(self):
    avg = self.getAvg() 
    if avg >= 90:
      grade = "A"
    elif avg >= 85:
      grade = "B"
    elif avg >= 80:
      grade = "C" 
    return grade   

  def __str__(self): #print()함수를 통해 객체의 정보를 출력 자동호출
    msg = f'{self.name}\t{self.getSum()}\t{self.getAvg()}\t{self.getGrade()}'
    return msg
```

```python
filename = '/content/score.txt'
f = open(filename, encoding = 'utf-8')
data = f.readlines()
df=[]
for line in data:
  data1 = line.rstrip("\n").split(',')
  for factor in range(1,len(data1)):
    data1[factor] = int(data1[factor])
  df.append(data1)

print("이름", "총점", "평균", "학점", sep="\t") 


for i in range(len(df)):
  s1 = Student(df[i][0], df[i][1], df[i][2], df[i][3] )
  print(str(s1))
```

```python
class Member:
  def __init__(self, mId, mPw):
    self.mId = mId
    self.mPw = mPw

  def __str__(self):
    result = f'ID : {self.mId} \n PW : {self.mPw}'
    return result
```

```python
members = {}

class MemberManage(Member):
  def __init__(self, mId, mPw):
    super().__init__(mId, mPw)

  def addMember(self):
    
    if type(members) is not dict:
      print("딕셔너리 형식으로 입력하세요! 입력한 데이터 타입:", type(members))
    elif self.mId == '' or self.mPw == '':
      print("아이디와 비밀번호를 입력하세요")
    else:
        if self.mId in members:
            print(f"{self.mId}가 계정에 이미 있습니다. 가입이 불가능합니다")
        else:
            members[self.mId] = self.mPw
            print(self.mId,"계정등록 추가 완료")
                
  def printMember(self):
      return f'ID: {self.mId} \n PW : {self.mPw}'

 
    
```

```python
#메인코드
m = MemberManage()
m.addMember(Member('hee8460', '1234'))
m.addMember(Member('hee8461', '1234'))
m.addMember(Member('hee8462', '1234'))
```
